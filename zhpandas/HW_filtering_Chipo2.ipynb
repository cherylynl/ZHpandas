### Step 3. Assign it to a variable called chipo.

```python
url= 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo =  pd.read_csv(url, sep = '\t')
```

### Step 4. How many products cost more than $10.00?

```python

#price_change = [float(value[1 : -1]) for value in chipo.item_price]
#chipo.item_price= price_change
#chipo.dropna(axis=1)
#find products over $10.00
#chipo.item_price.gt('10.00')
#chipo.loc[chipo['item_price'] >=10.00]
#print(chipo['item_name'].where(chipo['item_price'] < 10))

#change item_price to float
chipo.astype({'item_price': 'float'}).dtypes

# eliminate missing costs
chipo.dropna(axis = 1)

#find products over $10.00
filtered_sales = chipo[chipo['item_price']>=10.00] 
filtered_sales.drop_duplicates(subset = ['item_name', 'item_price'])
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.dropna(inplace = True)
chipo_nodups = chipo.drop_duplicates(subset=['item_name','item_price'])

chipo_nodups.loc[:, "item_name": "item_price"]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by = 'item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
expensive = chipo.sort_values(by = "item_price", ascending = False)
expensive.head(1)

```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
VSB = chipo['item_name'] == 'Veggie Salad Bowl'
chipo[VSB].value_counts().sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
filt = (chipo['item_name'] == 'Canned Soda') & (chipo['quantity']>= 2 )
chipo[filt].value_counts().sum()
```
