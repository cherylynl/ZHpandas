### Step 3. Select the Apple, Tesla, Twitter, IBM, LinkedIn stocks symbols and assign them to a variable called stocks

```python
stocks = ["AAPL", "TSLA", "TWTR", "IBM"]# LinkedIn ("LNKD"), no longer traded
```

### Step 4. Read the data from google, assign to df and print it

```python
# df = data.DataReader(stocks, 'google', start, end) - google not implemented for datareader using Yahoo instead

df = web.DataReader(stocks, data_source='yahoo', start= start) 
df
```
